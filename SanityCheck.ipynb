```python
%load_ext autoreload
%autoreload 2

import adl_theory as adl
import numpy as np
```

```python
phi = [0.8]
beta = [0.5, 0.3]

results = adl.adl_dynamic_effects(phi, beta)

# Display
results["summary"]
results["table"].head()

# Tests
np.testing.assert_allclose(
    results["summary"]["long_run_effect"],
    4.0
)

np.testing.assert_allclose(
    results["table"]["Dynamic Effect"].iloc[0],
    0.5
)

np.testing.assert_allclose(
    results["table"]["Dynamic Effect"].iloc[1],
    0.7
)

np.testing.assert_allclose(
    results["table"]["Dynamic Effect"].iloc[2],
    0.56
)
```
